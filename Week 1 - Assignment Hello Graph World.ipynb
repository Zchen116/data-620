```python
import matplotlib.pyplot as plt
import networkx as nx
```

```python
%matplotlib inline
G = nx.Graph()

G.add_edges_from([("Andre","Carol"),
                  ("Andre","Diane"),
                  ("Andre","Beverly"),
                  ("Andre","Fernando"),
                  ("Beverly","Diane"),
                  ("Beverly","Ed"),
                  ("Beverly","Garth"),
                  ("Carol","Diane"),
                  ("Carol","Fernando"),
                  ("Diane","Ed"),
                  ("Diane","Fernando"),
                  ("Diane","Garth"),
                  ("Ed","Garth"),
                  ("Fernando","Garth"),
                  ("Fernando","Heather"),
                  ("Garth","Heather"),
                  ("Heather","Ike"),
                  ("Ike","Jane")])

list(G.degree)[10:20]
```

```python
nx.draw(G, with_labels = True, font_weight="bold", font_size=15)

plt.axis('off')
plt.show() #display
```
